```python
non_smoker_df = df[df.smoker == "no"]
non_smoker_df.head()
```

```python
plt.title("Age vs Charges")
sns.scatterplot(data=non_smoker_df, x="age", y="charges")
plt.show()
```

```python
def estimate_charges(age, w, b):
  return w * age + b
```

```python
w = 50
b = 100
age = 30
estimate_charges(age, w, b)
```

```python
non_smoker_df.head()
non_smoker_df.info()
```

```python
ages = non_smoker_df.age
ages
```

```python
estimate_charges(ages, w, b)
```

```python
non_smoker_df["estimate_charges"] = estimate_charges(ages, w, b)
non_smoker_df.head()
```

```python
target = non_smoker_df.charges
plt.plot(ages, non_smoker_df.estimate_charges, "r", alpha=0.9)
plt.scatter(ages, target, alpha=0.8)
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Charges")
plt.legend(["estimate", " Actual"])
plt.show()
```

```python
def try_parrams(w, b):
  ages = non_smoker_df.age
  target = non_smoker_df.charges
  estimate_charged = estimate_charges(ages, w, b)

  plt.plot(ages, estimate_charged, "r", alpha=0.9)
  plt.scatter(ages, target, s=8, alpha=0.8)
  plt.xlabel("Age")
  plt.ylabel("Charges")
  plt.title("Age vs Charges")
  plt.legend(["estimate", " Actual"])
  plt.show()
```

```python
try_parrams(60, 200)
```

```python
try_parrams(400, 5000)
```

```python
try_parrams(270, -4000)
```

```python
import numpy as np
def rmse(target, estimate):
  return np.sqrt(np.mean((target - estimate) ** 2))
```

```python
w = 50
b = 100
try_parrams(w, b)
```

```python
targets = non_smoker_df.charges
predicted = estimate_charges(non_smoker_df.age, w, b)
rmse(targets, predicted)
```

```python
targets = non_smoker_df.charges
predicted = estimate_charges(non_smoker_df.age, 250, -4000)
rmse(targets, predicted)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model_input = non_smoker_df[["age"]]
target = non_smoker_df.charges
model.fit(model_input, target)
```

```python
model.predict(np.array([[30], [23], [64]]))
```

```python
model.predict(model_input)
```

```python
model_input
```

```python
target
```

```python
rmse(target, model.predict(model_input))
```

```python
model.coef_
```

```python
try_parrams(model.coef_, model.intercept_)
```
